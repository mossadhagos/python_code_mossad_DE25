```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
# generate some data
x = np.linspace(-5, 10, 100)
x
```

```python
# generate data f
f  = np.sin(x)
f
```

### the first approach - functional/matplotlib

```python
plt.figure()
plt.plot(x,f)
plt.title("sin(x)")
plt.xlabel("Y")
plt.ylabel("X")
```

### the scond approach - oop

```python
fig = plt.figure() # creat one figuree 
ax = plt.axes () # on the figure crate an axe
ax.plot(x, f) # on the axe, create a plot 
ax.set(xlim= (-10,10), ylim=(-5,5), xlabel="x", ylabel="y") # continue modify the axe object in difrenen aspectes 
```
